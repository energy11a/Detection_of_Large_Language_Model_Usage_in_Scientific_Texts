```python
import pandas as pd
import json
import re

# Eesti ja Briti nimede tüüpilised mustrid
estonian_pattern = re.compile(r"\b(Aleksandr|Sergei|Vladimir|Andrei|Martin|Aleksei|Andres|Dmitri|Igor|Toomas|Viktor|Margus|Roman|Oleg|Jüri|Kristjan|Maksim|Urmas|Rein|Aivar|Alexander|Marko|Artur|Jaan|Ivan|Nikolai|Tarmo|Jevgeni|Peeter|Indrek|Sander|Priit|Juri|Meelis|Mihhail|Nikita|Aleksander|Andrus|Rasmus|Jaanus|Raivo|Pavel|Siim|Marek|Oliver|Mati|Valeri|Mihkel|Artjom|Tiit|Tõnu|Sergey|Anton|Markus|Oleksandr|Karl|Tanel|Robert|Madis|Eduard|Taavi|Janek|Mark|Lauri|Kaspar|Kevin|Mart|Erik|Daniil|Kirill|Kaido|Vadim|Kalev|Raul|Vladislav|Vitali|Konstantin|Ilja|Kalle|Ants|Silver|Daniel|Anatoli|Kristo|Henri|Robin|Vello|Aare|Ain|Vjatšeslav|Stanislav|Denis|Tõnis|Toivo|Arvo|Oskar|Rainer|Ruslan|Jaak|Rain|Olga|Irina|Jelena|Svetlana|Tatjana|Anna|Galina|Valentina|Natalja|Maria|Marina|Katrin|Tiina|Sirje|Anne|Jekaterina|Natalia|Julia|Kristina|Ülle|Tiiu|Ene|Diana|Viktoria|Piret|Anneli|Nadežda|Aleksandra|Laura|Riina|Kadri|Kristi|Merike|Tamara|Anu|Ljudmila|Marika|Niina|Maie|Elena|Eve|Tatiana|Oksana|Kristiina|Mare|Triin|Annika|Jana|Reet|Karin|Kersti|Malle|Merle|Alina|Liina|Sofia|Veronika|Ljubov|Sandra|Pille|Helen|Kristel|Liis|Inna|Margit|Larissa|Kaja|Heli|Alla|Ljudmilla|Urve|Tiia|Lea|Eha|Anastassia|Maire|Eva|Veera|Linda|Külli|Nina|Helle|Polina|Anastasia|Ingrid|Aino|Liudmila|Kätlin|Darja|Mari|Maarja|Margarita|Aili|Krista|Hanna|Maris|Karina|Monika|Liisa|Kadi)\b", re.IGNORECASE)
british_pattern = re.compile(r"\b(Oliver|George|Noah|Arthur|Muhammad|Leo|Oscar|Harry|Archie|Henry|Theodore|Jack|Charlie|Freddie|Alfie|Jacob|Thomas|Isaac|Finley|Lucas|Alexander|William|Arlo|Roman|Jude|Elijah|James|Reuben|Edward|Hugo|Sebastian|Louis|Mason|Rory|Benjamin|Ethan|Albie|Ezra|Joseph|Max|Adam|Tobias|Samuel|Finn|Jesse|Daniel|Caleb|Jaxon|Harrison|Zachary|Rowan|Hugo|Ellis|Ronnie|Louis|Kit|Myles|Milo|Jasper|Grayson|Albert|Blake|Ellis|Felix|Otis|Leon|Sonny|Rupert|Bodhi|Frankie|Theo|Tobias|Zach|Monty|Reggie|Wilfred|Hunter|Darcy|Lennie|Sonny|Ellis|Brodie|Louie|Rex|Barney|Kit|Alfred|Basil|Heath|Laurie|Digby|Clyde|Vernon|Dougie|Hugh|Basil|Percy|Bertie|Clive|Neville|Olivia|Amelia|Isla|Ava|Lily|Ivy|Freya|Florence|Mia|Willow|Sophia|Rosie|Ella|Grace|Evie|Charlotte|Sienna|Emily|Daisy|Poppy|Elsie|Sofia|Millie|Phoebe|Aria|Matilda|Alice|Harper|Esme|Maya|Isabelle|Ada|Bonnie|Imogen|Evelyn|Luna|Eliza|Violet|Thea|Lottie|Arabella|Penelope|Aurora|Erin|Chloe|Zara|Bella|Lyra|Rose|Nancy|Clara|Orla|Edith|Hallie|Ellie|Robyn|Amber|Mabel|Gracie|Annie|Delilah|Nova|Lola|Harriet|Olive|Margot|Francesca|Summer|Niamh|Ayla|Heidi|Beatrice|Indie|Lucia|Anya|Talia|Quinn|Skye|Maya|Lola|Elodie|Cora|Lyla|Maeve|Wren|Isla|Florence|Jemima|Aisha|Zainab|Lila|Ottilie|Indigo|Faith|Betsy|Seren|Marlowe|Anouk|Georgina|Winnie)\b", re.IGNORECASE)

def is_estonian_or_british(author_list):
    """Kontrollib, kas mõni autori nimi viitab Eesti või Briti päritolule."""
    for author in author_list:
        name = author.get("name", "")
        if not isinstance(name, str):
            continue  # vahele, kui pole string
        if estonian_pattern.search(name) or british_pattern.search(name):
            return True
    return False

# Loeme failist andmed tükiti
filtered_rows = []
chunks = pd.read_json("DBLP-Citation-network-V18.jsonl", lines=True, chunksize=10000)

for chunk in chunks:
    # Eemalda kõik kirjed enne 2016
    chunk = chunk[chunk["year"] >= 2016]
    # Filtreeri read, mille autorinimed sobivad mustriga
    mask = chunk["authors"].apply(is_estonian_or_british)
    filtered_rows.append(chunk[mask])

# Ühenda kõik sobivad read üheks DataFrame’iks
result_df = pd.concat(filtered_rows, ignore_index=True)

# Vaatame tulemusi
print(result_df[["title", "year", "authors"]].head())

# salvestab CSVks
result_df.to_csv("estonian_or_british_authors.csv", index=False)
```

```python
def is_from_estonia(author_list):
    """Tagastab True, kui mõne autori 'org' sisaldab sõna 'Estonia'."""
    if not isinstance(author_list, list):
        return False
    author= author_list[0]
    org = author.get("org") if isinstance(author, dict) else None
    name = author.get("name", "")
    if isinstance(org, str) and "Estonia" in org and estonian_pattern.search(name):
        return True
    return False

# Rakenda filtreerimist
estonia_df = result_df[result_df["authors"].apply(is_from_estonia)]
#Salvesta DataFrame CSV-failiks
estonia_df.to_csv("estonia_df.csv", index=False)

# Kontroll
print(estonia_df[["title", "year", "authors"]].head())
print(f"Kokku artikleid Eesti asutuste autoritega: {len(estonia_df)}")
```

```python
def is_from_uk(author_list):
    """Kontrollib, kas mõne autori 'org' väljal on UK/England/Great Britain."""
    author= author_list[0]
    org = author.get("org") if isinstance(author, dict) else None
    name = author.get("name", "")
    if not isinstance(org, str):
        return False
    org_lower = org.lower()
    if any(country in org_lower for country in ["united kingdom", "england", "great britain", "scotland", "wales", "northern ireland"]) and british_pattern.search(name):
        return True
    return False

# Rakenda filtreerimist
uk_df = result_df[result_df["authors"].apply(is_from_uk)]
#Salvesta DataFrame CSV-failiks
uk_df.to_csv("uk_df.csv", index=False)

# Kontroll
print(uk_df[["title", "year", "authors"]].head())
print(f"Kokku artikleid Inglise asutuste autoritega: {len(uk_df)}")
```
