```python
import os
from pathlib import Path
import re
import random

# alamkaustad
subfolders = ["MAE", "SAE", "SAU", "MAU"]

# outputs kaust
output_dir = Path("proccessed")
output_dir.mkdir(exist_ok=True)

# strateegia nimi
strateegia = "masked_block_x_div_4"

# regex lauselõpu jaoks
SENTENCE_SPLIT_REGEX = re.compile(r'(?<=[.!?])(?=\s+[A-Za-zÕÄÖÜõäöüŠšŽž])')

def mask_sentence(sentence: str) -> str:
    """Asendab kõik märgid '_'-ga, säilitades pikkuse."""
    return re.sub(r'.', '_', sentence)

for sub in subfolders:
    sub_path = Path(sub)
    if not sub_path.exists():
        print(f"Kaust puudub: {sub}")
        continue

    for file in sub_path.glob("*.txt"):
        original_name = file.stem
        new_name = f"{sub}_{original_name}_{strateegia}.txt"
        target = output_dir / new_name

        with open(file, "r", encoding="utf-8") as f:
            text = f.read()

        # jaga tekst lauseteks
        sentences = SENTENCE_SPLIT_REGEX.split(text)

        x = len(sentences)
        if x == 0:
            continue

        mask_count = max(1, x // 4)

        # juhuslik algus
        start_idx = random.randint(0, x - mask_count)

        processed_sentences = []
        for i, sentence in enumerate(sentences):
            if start_idx <= i < start_idx + mask_count:
                processed_sentences.append(mask_sentence(sentence))
            else:
                processed_sentences.append(sentence)

        # ühenda tagasi
        processed_text = "".join(processed_sentences)

        with open(target, "w", encoding="utf-8") as f:
            f.write(processed_text)

        print(f"Maskitud {mask_count} lauset alates indeksist {start_idx}: {file} -> {target}")
```

```python
import re

SENTENCE_SPLIT_REGEX = re.compile(r'(?<=[.!?])\s+(?=[A-ZÕÄÖÜŠŽ])')
PARA_THRESHOLD = 0.6

# ----------------------------------------
# 1. Märgib teksti sees, kus lause algab uuelt realt
# ----------------------------------------
def mark_sentence_starts_in_text(text: str):
    """
    Lisab teksti iga lause algusesse märgi <<SENTENCE_START>>
    """
    text = re.sub(r'(\.) ([A-Z])', r'\1\n\2', text)
    return text

# ----------------------------------------
# 3. Parandab poolikud sõnad ja lause sees reavahetused
# ----------------------------------------
def fix_text_inside(text: str):
    """
    Eemalda poolitatud sõnad ja lause sees olevad reavahetused teksti sees.
    """
    # 1. Poolitatud sõnad: sidekriips + reavahetus -> liidab sõna
    text = re.sub(r'(\w)-\n(\w)', r'\1\2', text)
    # 2. Ülejäänud reavahetused lause sees → tühik
    text = re.sub(r'([a-z])\n(\w)', r'\1 \2', text)
    # 3. Topelt tühikud
    #text = re.sub(r'\s{2,}', ' ', text)
    return text


def split_text_with_paragraphs(text: str) -> str:
    """
    Jagab teksti ridadeks ja lisab paragrahvi tühja rea
    juhul, kui praegune rida algab suure tähega ja eelmine lõppeb väiketähe ja punktiga.
    """
    # Eemaldame liigsed reavahetused ja poolitatud sõnad juba varem (soovitavalt fix_text_inside)
    #text = text.strip()
    
    # Jagame tekst esialgu reavahetuse järgi
    raw_lines = text.split('\n')
    
    final_lines = []
    prev_line = ""

    for line in raw_lines:
        line = line.strip()
        if not line:
            continue  # ignoreeri tühje ridu

        # Paragrahvi tuvastus
        if prev_line:
            # kontroll: eelmine lõppeb väiketähe ja punktiga
            if re.match(r'.*[a-z]\.$', prev_line) and re.match(r'^[A-ZÕÄÖÜŠŽ]', line) and len(prev_line)<len(line)*0.6:
                final_lines.append("")  # lisa tühja rea markerina paragrahv
        final_lines.append(line)
        prev_line = line

    return "\n".join(final_lines)
```

```python
def process_file(file_path, output_dir):
    with open(file_path, "r", encoding="utf-8") as f:
        text = f.read()
    # 2. Märgime paragrahvid teksti sees
    text = split_text_with_paragraphs(text)
    # 3. Parandame poolikud sõnad ja reavahetused
    text = fix_text_inside(text)
    # 1. Märgime lause algused
    text = mark_sentence_starts_in_text(text)
    
    
    # Salvestame faili
    output_dir.mkdir(exist_ok=True)
    target = output_dir / file_path.name
    with open(target, "w", encoding="utf-8") as f:
        f.write(text)

    print(f"Töödeldud: {file_path} -> {target}")

subfolders = ["MAE", "SAE", "SAU", "MAU"]

for sub in subfolders:
    sub_path = Path(sub)
    if not sub_path.exists():
        print(f"Kaust puudub: {sub}")
        continue

    output_dir = Path(f"{sub}_norm")

    for file in sub_path.glob("*.txt"):
        process_file(file, output_dir)
```

```python
#!pip install langchain_core
#!pip install langchain_openai
#Setup LLm client
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import PromptTemplate
from langchain_openai import ChatOpenAI
from os import getenv
from langchain_huggingface import HuggingFaceEndpoint, ChatHuggingFace

#from dotenv import load_dotenv
#	GPT-4 (omni , omni-mini) ükski ei sobi andmepoliitikaga
#	Gemini
#	Mistral large
#	Llama-3
#	Cohere
#	Mistral-Nemo
# openai/gpt-oss-120b:free ei sobi
# google/gemma-3-27b-it:free ükski ei sobi
#deepseek/deepseek-r1-0528:free  vajab täpsemaid juhiseid kuna tahab lisada formaatimist ja lisa lauseid

OPENROUTER_API_KEY = "sk-or-v1-"
OPENROUTER_API_KEY = "sk-or-v1-"
OPENROUTER_BASE_URL = "https://openrouter.ai/api/v1"
HF_API_KEY = ""
def setup_llm_client(provider="openrouter", model=None):
    if provider == "openrouter":
        from langchain_openai import ChatOpenAI

        return ChatOpenAI(
            openai_api_key=OPENROUTER_API_KEY,
            openai_api_base="https://openrouter.ai/api/v1",
            model=model or "meta-llama/llama-3.3-70b-instruct:free",
            temperature=0.2
        )

    elif provider == "google":
        from langchain_google_genai import ChatGoogleGenerativeAI

        return ChatGoogleGenerativeAI(
            model=model or "gemini-flash-latest",
            google_api_key="",
            temperature=0.2
        )

    elif provider == "huggingface":
        from langchain_huggingface import HuggingFaceEndpoint
        from langchain_core.runnables import RunnableLambda

        # Loo LLM instants
        llm = HuggingFaceEndpoint(
            repo_id=model,
            huggingfacehub_api_token=HF_API_KEY,
            temperature=0.2,
            max_new_tokens=1024,
            task="text-generation",
        )

        # Wrapper mis töötab nii str kui PromptValue inputiga
        def hf_invoke(x):
            # Kui str, kasuta otse; kui PromptValue, konverteeri stringiks
            if isinstance(x, str):
                prompt_str = x
            else:
                prompt_str = x.to_string()

            result = llm.invoke(prompt_str)

            # tagasta alati string
            if result is None:
                return ""
            if isinstance(result, str):
                return result.strip()
            if isinstance(result, dict):
                return result.get("generated_text", "").strip()
            return str(result).strip()

        return RunnableLambda(hf_invoke)

    else:
        raise ValueError("Unsupported provider")
```

```python
def load_template(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        return file.read()

def create_prompt_template(template_content):
    return PromptTemplate(
        input_variables=['article_content'], #pass input variable into template
        template=template_content
    )
```

```python
import os
import time
import re


def safe_invoke(chain, input_data, max_retries=3):

    for attempt in range(max_retries):
        try:
            return chain.invoke(input_data)

        except Exception as e:

            msg = str(e)

            # Handle Gemini / API rate limits
            if "RESOURCE_EXHAUSTED" in msg or "429" in msg:

                match = re.search(r"retry in ([0-9.]+)s", msg)

                if match:
                    wait_time = float(match.group(1)) + 1
                else:
                    wait_time = 30

                print(f"Rate limit hit. Waiting {wait_time:.1f}s before retry...")
                time.sleep(wait_time)

            else:
                raise e

    raise RuntimeError("Max retries exceeded")


def process_directory(indirectory_path, outdirectory_path, chain):

    os.makedirs(outdirectory_path, exist_ok=True)

    for filename in os.listdir(indirectory_path):

        if not filename.endswith(".txt") or filename.endswith("_out.txt"):
            continue

        input_path = os.path.join(indirectory_path, filename)

        output_filename = filename.replace(".txt", "_out.txt")
        output_path = os.path.join(outdirectory_path, output_filename)

        # Skip if output already exists
        if os.path.exists(output_path):
            print(f"Skipping {filename} (output already exists)")
            continue

        print(f"Processing file: {filename}")

        try:

            article_content = read_file(input_path)

            input_data = {
                "article_content": article_content
            }

            output = safe_invoke(chain, input_data)

            save_output(output, output_path)

            print(f"Output saved: {output_filename}")

            # small delay to reduce rate-limit probability
            time.sleep(2)

        except Exception as e:

            print(f"Error processing {filename}: {e}")

            # wait before next file
            time.sleep(10)


def read_file(filepath):
    with open(filepath, "r", encoding="utf-8") as f:
        return f.read()


def save_output(text, filepath):
    with open(filepath, "w", encoding="utf-8") as f:
        f.write(text)
```

```python
import os

def process_directory_pairs(indirectory_path, outdirectory_path, chain):
    os.makedirs(outdirectory_path, exist_ok=True)

    files = [f for f in os.listdir(indirectory_path) if f.endswith(".txt")]

    #failid baasinime järgi gruppi
    pairs = {}

    for filename in files:
        if filename.endswith("_masked_row0_out.txt"):
            base = filename.replace("_masked_row0_out.txt", "")
            pairs.setdefault(base, {})["row0"] = filename

        elif filename.endswith("_masked_row1_out.txt"):
            base = filename.replace("_masked_row1_out.txt", "")
            pairs.setdefault(base, {})["row1"] = filename

    for base, pair in pairs.items():
        if "row0" not in pair or "row1" not in pair:
            print(f"Paar puudulik, jäetakse vahele: {base}")
            continue

        file0 = os.path.join(indirectory_path, pair["row0"])
        file1 = os.path.join(indirectory_path, pair["row1"])

        print(f"Processing pair: {pair['row0']} + {pair['row1']}")

        content0 = read_file(file0)
        content1 = read_file(file1)

        # input LLM-ile
        input_data = {
            "masked_row0": content0,
            "masked_row1": content1
        }

        output = chain.invoke(input_data)

        output_filename = f"{base}_out.txt"
        output_path = os.path.join(outdirectory_path, output_filename)
        save_output(output, output_path)

        print(f"Output saved: {output_filename}")
```
